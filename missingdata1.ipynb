```python
#loading dataset
dataset=pd.read_csv("/content/Placement_Dataset.csv")
#dataset shape
dataset.shape
```

```python
#Number of missing values
dataset.isnull().sum()
```

#Handling missing values using Imputation

```python
dataset.head()
```

```python
#Analyzing the distribution of data in the salary column
fig,ax=plt.subplots(figsize=(8,8))
sns.histplot(dataset["salary"])
```

```python
#Replacing missing values with median value
dataset["salary"]=dataset["salary"].fillna(dataset["salary"].median())
```

```python
#Using Dropping
salary_dataset=pd.read_csv("/content/Placement_Dataset.csv")
```

```python
#shape of the dataset
salary_dataset.shape
```

```python
#Number of missing values
salary_dataset.isnull().sum()
```

```python
#dropping the rows that has missing values
salary_dataset=salary_dataset.dropna(how="any")
```

```python
#Verifying the whether the missing values are dropped or not
salary_dataset.isnull().sum()
```
